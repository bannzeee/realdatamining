주성분 분석


```python
df.columns
```

```python
# 주성분 분석 라이브러리
from sklearn.decomposition import PCA

# 데이터 정규화 수행
# 앞차시 정규화 완성 데이터 사용
pca_col = ['등록금', '인문사회', '자연과학', '예체능', '공학']
dfp = df[pca_col]
dfp
```

```python
# PCA 모델 생성
pca = PCA(n_components=2) # 기존 5개의 특징을, 주성분 2개로 축소

# 주성분 분석 수행
X_pca = pca.fit_transform(dfp)

# PCA 결과를 데이터프레임으로 변환
df_pca = pd.DataFrame(data=X_pca, columns=['C1', 'C2'])
df_pca
```

```python
# PCA 열 결합
df = pd.concat([df, df_pca], axis=1)
df
```

```python
# 시각화를 통한 차원축소 이유 확인
sns.scatterplot(x='C1', y='C2', hue='설립구분', data=df)
plt.plot()
```

```python
# 데이터 나누기
df = pd.read_csv('/content/drive/MyDrive/SW_데이터마이닝/DATA/data_08_등록금_정제.csv')
df1 = df[df['설립구분'] == '사립']
df2 = df[df['설립구분'] == '국립']

sns.barplot(x='설립구분', y='등록금', data=df)
plt.title('사립 vs. 국립 등록금 평균')
plt.show()
```

```python
# 샤피로-윌크 검정 : 데이터 정규성 확인
import scipy.stats as stats

def shapiro(msg, ds) :
  print(msg)

  #샤피로-윌크 검정 수행
  stat, p_value = stats.shapiro(ds)

  #결과출력
  print('검정 통계량(statistic):', stat)
  print('p-값(p-value):', p_value)

  #유의수준 0.05에서 검정 결과 출력
  alpha = 0.05
  if p_value > alpha:
    print('데이터는 정규분포를 따른다')
  else:
    print('데이터는 정규분포를 따르지 않는다')
```

```python
# 정규성 검증
shapiro('사립학교 등록금 정규성 검증', df1['등록금'])
print('-'*50)
shapiro('국립학교 등록금 정규성 검증', df2['등록금'])
```

```python
df1['등록금'].mean()
```

```python
df2['등록금'].mean()
```

```python
# t-검정 (두 집단 간 비교)
t_stat, p_value = stats.ttest_ind(df1['등록금'], df2['등록금'])

print('등록금 평균')
print(f"사립 : {df1['등록금'].mean():.2f}, 국립 : {df2['등록금'].mean():.2f}")
print('t-검정 결과: ')
print('t-statistic: ', t_stat)
print('p-value: ', p_value)

#유의수준 0.05에서 검정 결과 출력
alpha = 0.05
if p_value > alpha:
  print('사립대와 국립대 평균의 차이는 없다')
else:
  print('사립대와 국립대 평균의 차이는 있다')
```
