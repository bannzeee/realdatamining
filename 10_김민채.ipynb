```python
#주성분 분석 라이브러리
from sklearn.decomposition import PCA

#데이터 정규화 수행
#앞차시 정규화 완성 데이터 사용
pca_col =  ['등록금', '인문사회', '자연과학', '예체능', '공학']
dfp = df[pca_col]
dfp
```

```python
#PCA 모델 생성
pca = PCA(n_components=2) #기존 5개의 특징을, 주성분 2개로 축소

#주성분 분석 수행
X_pca = pca.fit_transform(dfp)

#PCA 결과를 데이터프레임으로 변환
df_pca = pd.DataFrame(data = X_pca, columns = ['C1', 'C2'])
df_pca
```

```python
#PCA 열 결합
df = pd.concat([df,df_pca],axis=1)
df
```

```python
#데이터 나누기
df = pd.read_csv('/content/drive/MyDrive/데이터마이닝/DATA/data_08_등록금_정제.csv')
df1 = df[df['설립구분']=='사립']
df2 = df[df['설립구분']=='국립']

sns.barplot(x='설립구분', y='등록금', data=df)
plt.title('사립 vs 국립 등록금 평균')
plt.show()
```

```python
# 샤피로- 윌크 검정 : 데이터 정규성 확인
#외울필요는 없음
import scipy.stats as stats

def shapiro(msg,ds):
  print(msg)

  #샤피로-윌크 검정 수행
  stat,p_value = stats.shapiro(ds)

  #결과출력
  print('검정 통계량(statistic) : ',stat)
  print('p-값(p-value) : ',p_value)

  #유의수준 0.05에서 검정 결과 출력 (더 정밀하게 해야 하는 경우에는 99% 즉 0.01로 해야 함)
  #여기 결과 해석 중요!!!!!!!!!!!!!!!!!!!!!!!!!!!

  alpha = 0.05
  if p_value > alpha: #귀무가설(=,집단간에차이가 없다)을 설정함에 있어서 정규분포로 설정함 기준: 0.05보다 크다 -> 정규분포로 해석함
    print('데이터는 정규분포를 따른다.')
  else:
    print('데이터는 정규분포를 따르지 않는다.')
```

```python
#정규성 검증
shapiro('사립학교 등록금 정규성 검증',df1['등록금'])
print('-'*50)
shapiro('국립학교 등록금 정규성 검증',df2['등록금'])
```

-> statistics 무시, p값만 생각하기

---
p > 0.05 -> 정규분포이다


```python
df1['등록금'].mean()
```

```python
df2['등록금'].mean()
```

```python
#t-검정 (두 집단 간 비교)
t_stat, p_value = stats.ttest_ind(df1['등록금'],df2['등록금']) #ttest_ind 중요

print('등록금 평균')
print(f"사립 : {df1['등록금'].mean():.2f}, 국립 : {df2['등록금'].mean():.2f}") #f스트링 형태 중요
print('t-검정 결과')
print('t- statistic : ',t_stat)
print('p-value : ',p_value)

#유의수준 0.05에서 검정 결과 출력
alpha = 0.05
if p_value > alpha:
  print('사립대와 국립대 평균의 차이가 없다.')
else:
  print('사립대와 국립대 평균의 차이가 있다.')
```

결론 해석이 가장 중요

---
t검정: 두 집단 비교 / A검정 : 세 집단 이상인 경우

---
이 두 집단이 차이가 있는지 없는지를 확인 -> t값과 p값이 유의미하게 분석될 때 차이가 있다고 해석 24.7은 몰라도 p값이 9.7에 e-49 -> 왼쪽으로 0.000.... -> 0.05보다 작다 -> 귀무가설(국립대와 사립대는 같다)을 기각하고 대립가설(국립대와 사립대는 같지 않다)을 채택한다 = 결론 : " 통계학적으로 유의미한 차이가 있다 "

---
전제 : 95% 신뢰수준 & 유의수준 알파 = 0.05
