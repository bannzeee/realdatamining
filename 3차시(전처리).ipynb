```python
## 조건에 부합되는 자료만 추출
# 역대 등록금 분석을 위하여, 의대 등록금이 있는 학교만 추출
dfm = df[df['의학'] !=0] #0이 아닌, 즉 df['의학']에 데이터가 있는 경우만 추출해서 df를 다시 만들어라
dfm
```

```python
#필요한 열만 추출 [[]] 리스트의 기능 : 대괄호 안에 대괄호
dfm = dfm[['학교','설립구분','지역','등록금','의학']]
dfm
```

```python
#필요한 열만 추출
dfm = dfm[['학교','설립구분','지역','등록금','의학']]

#행 인덱스 재정의
dfm.reset_index(inplace=True, drop= True)

#데이터프레임 확인
dfm
```

```python
#학교 등록금 대비 의대 등록금의 높은 싱위 10개 학교 조회
dfm['학교등록금대비'] = dfm['의학'] / dfm['등록금']
dfm
```

```python
#소수 한자리만 표기
dfm['학교등록금대비'] = round(dfm['의학'] / dfm['등록금'],1)
dfm
```

```python
#값 기준 (학교 등록금 대비),오름차순 sort
dfm['학교등록금대비'] =  round(dfm['의학'] / dfm['등록금'],1)
dfm.sort_values('학교등록금대비')
```
