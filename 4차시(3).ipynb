```python
## 런타임 다시 시작 후 진행
import matplotlib.pyplot as plt
plt.rc('font', family='NanumBarunGothic')

# 확인
plt.title('한글')
```

```python
# 파일 불러오기
import pandas as pd
df = pd.read_excel('/content/drive/MyDrive/SW_데이터마이닝/DATA/03_등록금_final.xlsx')
df
```

```python
# 지역을 행 인덱스로 변환 후, 원본 반영
df.set_index('지역', inplace=True)
df
```

```python
# 판다스 프레임 기본 그래프
df.plot(figsize=(8,4))
plt.show()
```

```python
df.plot(figsize=(8,4), marker='o') #영문o
plt.title('지역별 의대수')
plt.show()
```

```python
df.plot(figsize=(8,4), marker='o') #영문o
plt.title('지역별 의대수')
plt.xlabel('')
plt.xticks(range(len(df.index)))
plt.show()
```

```python
df.plot(figsize=(8,4), marker='o') #영문o
plt.title('지역별 의대수')
plt.xlabel('')
plt.xticks(range(len(df.index)), labels=df.index)
plt.show()
```

```python
# 내림차순 정렬
df.sort_values('합계', ascending=False)
```

```python
# 전체 데이터프레임 기본 그래프
df.sort_values('합계', ascending=False).plot(kind='bar', figsize=(8,4))
plt.show()
```

```python
df['합계'] #시리즈
```

```python
df[['합계']] #list 데이터프레임
```

```python
# 합계만, plot은 시리즈가 아닌, 데이터프레임으로 설정하기위하여, list의 [] 사용
df[['합계']].sort_values('합계', ascending=False).plot(kind='bar', figsize=(8,4))
plt.show()
```

```python
df[['합계', '국립']].sort_values('합계', ascending=False).plot(kind='bar', figsize=(8,4))
plt.show()
```

```python
# 판다스 막대그래프 : 시리즈도 가능
df['합계'].plot(kind='pie', autopct='%.1f%%')
plt.show()
```

```python
df['합계'].plot(kind='pie', autopct='%.1f%%')
plt.ylabel('') #합계 삭제
plt.title('지역별 의대수')
plt.show()
```

```python
# 히스토그램
df.plot(kind='hist', figsize=(8,4))
plt.title('지역별 의대수')
plt.show()
```

```python
# 박스플롯
df.plot(kind='box', figsize=(8,4))
plt.title('지역별 의대수')
plt.show()
```

```python
# 합계 삭제, loc 사용
df.loc[:, '국립':'사립'].plot(kind='box', figsize=(8,4)) #데이터프레임 설정, plot옵션 순
plt.title('지역별 의대수')
plt.show()
```
