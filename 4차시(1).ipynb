```python
#스타일 지정
plt.figure(figsize=(8,4))
plt.plot(df['지역'],df['합계'],'go--')
plt.show()
```

```python
#타이틀 표시
plt.figure(figsize=(8,4))
plt.plot(df['지역'],df['합계'],'go--')
plt.title('지역별 의대수')
plt.xlabel('지역')
plt.ylabel('학교수')
plt.xticks(rotation=45) #글자 각도 기울기
plt.show()
```

```python
plt.figure(figsize=(8,4))
plt.plot(df['지역'],df['합계'],'go--')
plt.title('지역별 의대수')
plt.xlabel('지역')
plt.ylabel('학교수')
plt.xticks(rotation=45) #글자 각도 기울기

#수치값을 화면에 표시
for i, v in enumerate(df['합계']) : #반복 가능한 객체를 입려으로 받아, 각 요소와 해댜ㅏㅇ 요소으ㅏ 안댁수룰
  plt.text(df['지역'][i],v,str(v))

plt.show()
```

enumerate 개별적으로 합계를 가져와서 i와 v(value)값으로 반복해서 찍어달라

```python
plt.figure(figsize=(8,4))
plt.plot(df['지역'],df['합계'],'go--')
plt.title('지역별 의대수')
plt.xlabel('지역')
plt.ylabel('학교수')
plt.xticks(rotation=90) #글자 각도 기울기

#수치값을 화면에 표시
for i, v in enumerate(df['합계']) :
  plt.text(df['지역'][i],v,str(v),ha='center',va='bottom')

plt.show()
```

```python
#여러가지 그래프
x = df['지역']
y = df['합계']

plt.figure(figsize=(10,10))

plt.subplot(3,2,1) #3행,2열,첫번째
plt.bar(x,y) #반드시x정의해줌
plt.xticks(rotation=90)
plt.title('막대그래프')

plt.show()
```

```python
#여러가지 그래프
x = df['지역']
y = df['합계']

plt.figure(figsize=(10,10))

plt.subplot(3,2,1) #3행,2열,첫번째 #subplot 에 s붙이면 안됨
plt.bar(x,y) #반드시x정의해줌
plt.xticks(rotation=90)
plt.title('막대그래프')

plt.subplot(3,2,2)
plt.barh(x,y) #barh
plt.title('가로막대그래프')

plt.show()
```
