```python
df2 = pd.read_excel('/content/drive/MyDrive/SW_데이터마이닝/DATA/03_의대등록금.xlsx')
dfg = df2.groupby(['지역'])[['등록금', '의학', '학교등록금대비']].mean() #그룹핑후, 지역이 index화됨
dfg
```

```python
# 그래프 생성
fig, ax1 = plt.subplots(figsize=(8, 4))

# 주축에 그래프 그리기
ax1.bar(dfg.index, dfg['의학'], color='yellow',label='의대등록금')  # dfg.index 가 '지역'의미

plt.show()
```

```python
# 그래프 생성
fig, ax1 = plt.subplots(figsize=(8, 4))

# 주축에 그래프 그리기
ax1.bar(dfg.index, dfg['의학'], color='yellow',label='의대등록금')

# 보조축 그래프 그리기
ax1.plot(dfg.index, dfg['학교등록금대비'], 'go--', label='학교등록금대비')

plt.show()
```

```python
# 그래프 생성
fig, ax1 = plt.subplots(figsize=(8, 4))

# 주축에 그래프 그리기
ax1.bar(dfg.index, dfg['의학'], color='yellow',label='의대등록금')

# 보조축 그래프 그리기
ax2 = ax1.twinx()  # ax1의 x축과 동일하게 사용
ax2.plot(dfg.index, dfg['학교등록금대비'], 'go--', label='학교등록금대비')

plt.show()
```

```python
# 그래프 생성
fig, ax1 = plt.subplots(figsize=(8, 4))

# 주축에 그래프 그리기
ax1.bar(dfg.index, dfg['의학'], color='yellow',label='의대등록금')

# 보조축 그래프 그리기
ax2 = ax1.twinx()
ax2.plot(dfg.index, dfg['학교등록금대비'], 'go--', label='학교등록금대비')

# 범례 추가
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()

ax2.legend(lines + lines2, labels + labels2, loc='upper left')
#외우지말고 복사해서 써라

plt.title('지역별 의대 등록금 평균')
plt.show()
```

```python
# 버블차트 : 산점도 그래프 응용
plt.scatter(df2['등록금'], df2['의학'])
plt.show()
```

```python
# 버블차트 : 산점도 그래프 응용
plt.scatter(df2['등록금'], df2['의학'],
           s=df2['학교등록금대비']*200,
            c=df2['학교등록금대비'], cmap='Spectral' #색상은 적당한것으로
          )
plt.show()
```

```python
# 버블차트 : 산점도 그래프 응용
plt.scatter(df2['등록금'], df2['의학'],
           s=df2['학교등록금대비']*200,
            c=df2['학교등록금대비'], alpha=0.5, cmap='Spectral' #명암 조정
          )
plt.colorbar()
plt.show()
```
