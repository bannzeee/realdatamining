```python
import pandas as pd
```

```python
df = pd.read_csv('/content/drive/MyDrive/SW_데이터마이닝/DATA/data_07_등록금_정규화_dfmm.csv')
df
```

```python
df.columns
```

```python
# 필요한 열 추출
df = df[['학교', '설립구분', '지역', '등록금', '인문사회', '자연과학', '예체능', '공학']]
df.head()
```

```python
dfc = df[['등록금', '인문사회', '자연과학', '예체능', '공학']]
dfc.corr()
```

```python
# 주성분 분석 라이브러리
from sklearn.decomposition import PCA

# 데이터 정규화 수행
# 정규화 완성 데이터 사용
pca_col = ['등록금', '인문사회', '자연과학', '예체능', '공학']
dfp = df[pca_col]
dfp
```

```python
# PCA 모델 생성
pca = PCA(n_components=2) # 기존 5개의 특징을, 주성분 2개로 축소

# 주성분 분석 수행
X_pca = pca.fit_transform(dfp)

# PCA 결과를 데이터프레임으로 변환
df_pca = pd.DataFrame(data=X_pca, columns=['C1', 'C2'])
df_pca
```

```python
# PCA 결과를 새로운 데이터프레임으로 변환
df_pca = pd.DataFrame(data=X_pca, columns=['C1', 'C2'])
df_pca['학교'] = df['학교']
df_pca['설립구분'] = df['설립구분']
df_pca['지역'] = df['지역']
df_pca = df_pca[['학교', '설립구분', '지역', 'C1', 'C2']]
df_pca
```

```python
# 시각화를 통한 차원축소 이유 확인
import seaborn as sns
import matplotlib.pyplot as plt
sns.scatterplot(x='C1', y='C2', hue='설립구분', data=df_pca)
plt.show()
```

```python
# 데이터 변환 : 타겟(국립,사립 예측), 범주형 데이터를 수치형 데이터로

# scikit-learn 라이브러리
from sklearn.preprocessing import LabelEncoder

# OrdinalEncoder 객체 생성
label_encoder = LabelEncoder()

df_pca['설립구분'] = label_encoder.fit_transform(df_pca['설립구분'])
df_pca.head()
```

```python
# train셋, test셋 나누기
from sklearn.model_selection import train_test_split
feature = df_pca.loc[:, ['C1', 'C2']]
target = df_pca['설립구분']

X_train, X_test, y_train, y_test = train_test_split(feature, target, shuffle=True, stratify = target, random_state=156)
```

```python
# 모델링 및 평가
from sklearn.metrics import accuracy_score, precision_score, recall_score
from sklearn.metrics import f1_score, confusion_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import roc_auc_score
```

```python
# 평가지표 함수
def get_clf_eval(y_test, pred):
  accuracy = accuracy_score(y_test, pred)
  precision = precision_score(y_test, pred)
  recall = recall_score(y_test, pred)
  f1 = f1_score(y_test, pred)
  auc = roc_auc_score(y_test, pred)
  return (accuracy, precision, recall, f1, auc) #튜플로 반환
```

```python
# 여러 가지 분류 모델을 한꺼번에 수행
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.gaussian_process import GaussianProcessClassifier
from sklearn.gaussian_process.kernels import RBF
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
```

```python
# 모델
model_dic = {
    "Nearest Neighbors" : KNeighborsClassifier(3),
    "SVM" : SVC(),
    "Linear SVM" : SVC(kernel='linear', C=0.025),
    "RBF SVM" : SVC(gamma=2, C=1),
    "Decision Tree" : DecisionTreeClassifier(),
    "Random Forest" : RandomForestClassifier(max_depth=5, n_estimators=10, max_features=1),
    "LogisticRegression" : LogisticRegression(),
    "Adaboost" : AdaBoostClassifier(),
    "Naive Bayes" : GaussianNB(),
    "QDA" : QuadraticDiscriminantAnalysis()
}
```

```python
model_p = {}
model_acc = {}

for key in model_dic.keys() :
  model = model_dic[key]
  print(f'{key} : training')
  model.fit(X_train, y_train)

  print(f'{key} : predict')
  model_p[key] = model.predict(X_test)
  model_acc[key] = get_clf_eval(y_test, model_p[key])
```

```python
model_acc
```

```python
# 데이터프레임으로 변환
model_acc = pd.DataFrame.from_dict(model_acc, orient='index',
                                   columns=['accuracy', 'precision', 'recall', 'F1', 'AUC'])
model_acc.sort_values('accuracy', ascending=False)
```

```python
from sklearn.metrics import RocCurveDisplay
disp = RocCurveDisplay.from_estimator(model_dic['Nearest Neighbors'], X_test, y_test)
for key in model_dic.keys() :
  model = model_dic[key]
  model.fit(X_train, y_train)

  # ROC 곡선 그리기
  RocCurveDisplay.from_estimator(model, X_test, y_test, ax=disp.ax_)
disp.figure_.suptitle("ROC curve comparision")

# 이정도 코딩만 알아두기 !! (범례를 바깥으로)
plt.legend(bbox_to_anchor=(1.05, 1)) #범례를 바깥으로
plt.show()
```
