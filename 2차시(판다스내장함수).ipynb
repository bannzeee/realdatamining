## 데이터프레임에서 행과 칼럼을 지정하여 데이터 보기

loc[]: 행 번호와 칼럼 이름을 사용해서 지정한 행과 칼럼이 가리키는 데이터를 가져옴



---
데이터프레임명.loc[행 번호,['칼럼 이름']]



```python
df.loc[1]
```

```python
df.loc[1,'판매량']
```

iloc[] : 0부터 시작한 행 순서값과 열 순서값을 이용해서 원하는 위치의 데이터를 가져온다

iloc[0부터 시작한 행 순서값, 0부터 시작한

칼럼 순서값]

```python
df.iloc[1,1]
```

**loc에서 0:3이라고 하면 0,1,2,3 모두를 불러오는거고 / 지정 안하면 모두 불러옴
iloc에서 0:3이라고 하면 0,1,2까지인거 / 지정안하면 모두 불러옴**

```python
df.loc[1:2, :]
```

```python
df.iloc[1:2, :]
```
