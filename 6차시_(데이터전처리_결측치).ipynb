```python
import pandas as pd
dfp = pd.read_excel('/content/drive/MyDrive/데이터마이닝/DATA/data_06_등록금_dfp.xlsx')
dfp
```

```python
#의학 대학이 없는 학교가 다수인 상태이므로 제외시킴
dfp = dfp.iloc[:,:-1] #제일마지막열제거
dfp
```

```python
#등록금을 내지 않는 학교(조회)
dfp[dfp['등록금'].isnull()]
```

```python
#등록금을 내지 않는 학교 삭제 : dropna 는 Nan 있는 모든 행 삭제됨 #subset 활용법 (수업시간에 다루지는 않을 것)
dfp.dropna(subset=['등록금'],inplace=True)
dfp
```

```python
# 삭제여부 확인
dfp.shape
```
