```python
# 전반적 표시 형태 변경
# 이 전체 코딩 중요 !!

fig, ax1 = plt.subplots(figsize=(10,7))
ax1.set_ylabel('출생아수 (천명)')

# y 축 범위 설정
ax1.set_ylim(200,600)

# y 축 tick(눈금) 값들을 리스트로 저장
ax1.set_yticks([200,300,400,500])

ax1.bar(df.index, df['출생아수'],color='#ff812d')

ax2 = ax1.twinx()

ax2.set_ylim(0,1.5)
ax2.set_yticks([0,1])

ax2.set_ylabel('합계출산율 (가임여성 1명당/명)')
ax2.plot(df.index, df['합계출산율'],color='#ffd100')
plt.show()
```

```python
# 범례

fig, ax1 = plt.subplots(figsize=(10,7))
ax1.set_ylabel('출생아수 (천명)')

ax1.set_ylim(200,600)
ax1.set_yticks([200,300,400,500])

ax1.bar(df.index, df['출생아수'],color='#ff812d')

# 범례
for idx, val in enumerate(df['출생아수']):
  ax1.text(idx, val+10, val, ha='center' )

ax2 = ax1.twinx()

ax2.set_ylim(0,1.5)
ax2.set_yticks([0,1])

ax2.set_ylabel('합계출산율 (가임여성 1명당/명)')
ax2.plot(df.index, df['합계출산율'],color='#ffd100')
plt.show()
```
